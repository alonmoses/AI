```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
oo = pd.read_csv('../data/olympics.csv',skiprows=4)
oo.head()
```

1.

```python
usmw = oo.groupby(['NOC', 'Medal', 'Edition', 'Gender'])
usmw
```

```python
usmw2 = usmw.agg({'Medal' : ['count']})
```

```python
usmw2 = usmw2.loc['USA','Gold']
usmw2
```

```python
usmw2.unstack().plot()
```

2.

```python
topAth = oo.groupby(['Athlete', 'Medal']).size()
```

```python
topAth = topAth.unstack(fill_value=0)
topAth
```

```python
topGoldAth = topAth.sort_values(['Gold','Silver', 'Bronze'], ascending=False)[['Gold','Silver', 'Bronze']].head(5)
```

```python
topGoldAth.plot(kind='bar')
```
