```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
oo = pd.read_csv('../data/olympics.csv', skiprows=4)
oo
```

## Plot Types

```python
tempoo = oo[oo.Edition == 1896]
```

```python
tempoo
```

```python
tempoo.Sport.value_counts().plot()
```

```python
tempoo.Sport.value_counts().plot(kind="bar")
```

```python
tempoo.Sport.value_counts().plot(kind="barh")
```

```python
tempoo.Sport.value_counts().plot(kind="pie")
```

## Plot colors

```python
tempoo.Sport.value_counts().plot(kind="bar", color='red')
```

```python
tempoo.Sport.value_counts().plot(color='GREEN')
```

## Figure Size

```python
tempoo.Sport.value_counts().plot(kind="bar", color='red', figsize=(40,30))
```

```python
tempoo.Sport.value_counts().plot(color='red', figsize=(5,5))
```

```python
tempoo.Sport.value_counts().plot(color='red', figsize=(10,10))
```

## Colormaps

```python
tempoo.Sport.value_counts().plot(kind='pie', colormap='Pastel1')
```

```python
tempoo.Sport.value_counts().plot(kind='pie', colormap='Pastel2')
```

```python
tempoo.Sport.value_counts().plot(kind='bar', colormap='Pastel1')
```
