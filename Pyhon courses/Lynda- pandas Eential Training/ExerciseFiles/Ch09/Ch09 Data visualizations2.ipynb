```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
oo = pd.read_csv('../data/olympics.csv', skiprows=4)
oo
```

```python
tempoo = oo[oo.Edition == 2008]
tempoo
```

```python
tempoo = tempoo.groupby(['NOC','Medal']).size().unstack('Medal', fill_value=0)
tempoo.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False)[['Gold', 'Silver', 'Bronze']]
```

## heatmap plot

```python
sns.heatmap(tempoo)
```

```python
plt.figure(figsize=(30,15))
sns.heatmap(tempoo.transpose())
```

## ColorMap plot

```python
tempoo = oo.groupby(['Athlete','Medal']).size().unstack('Medal', fill_value=0)
tempoo = tempoo.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False).head()[['Gold', 'Silver', 'Bronze']]
```

```python
tempoo.plot(kind='bar')
```
