```python
import pandas as pd
```

```python
oo = pd.read_csv('data/olympics.csv',skiprows=4)
oo
```

```python
athMedals = oo[oo.NOC == 'USA']
athMedals
```

```python
athMedals = athMedals.groupby(['Edition','Athlete','Medal']).size().unstack('Medal', fill_value=0)
athMedals['Total'] = athMedals['Gold'] + athMedals['Silver'] + athMedals['Bronze']
athMedals.reset_index(inplace=True)
```

```python
tu = [group.sort_values(by='Total', ascending=False)[:1] for year,group in athMedals.groupby('Edition')]
```

```python
top = pd.DataFrame()
for i in tu:
    top = top.append(i)

top.sort_values(by='Edition', ascending=False)
```
